show with python plots for bar chart, histogram, line diagram, pie diagram in a 2x2 subfigure matrix

```python
import matplotlib.pyplot as plt
import numpy as np

# Some data
x = np.arange(0., 10., 0.2)
y1 = np.sin(x)
y2 = np.cos(x)
categories = ['A', 'B', 'C', 'D', 'E']
values = [7, 15, 5, 20, 10]
hist_data = np.random.randn(100)

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Bar chart
axs[0, 0].bar(categories, values)  # Here, [0, 0] specifies the first row and first column of this grid.
axs[0, 0].set_title('Bar chart')

# Histogram
axs[0, 1].hist(hist_data, bins=20, color='green', alpha=0.7)
axs[0, 1].set_title('Histogram') #

# Line chart
axs[1, 0].plot(x, y1, label='sin(x)')
axs[1, 0].plot(x, y2, label='cos(x)', linestyle='--')
axs[1, 0].set_title('Line diagram')
axs[1, 0].legend()

# Pie chart
axs[1, 1].pie(values, labels=categories, autopct='%1.1f%%')
axs[1, 1].set_title('Pie chart')

# Adjust the spacing between the plots
plt.tight_layout()
plt.show()
```

Exercise 3: Count the k-permutations

Write a function that returns the number of all possible k-permutations of a list.

```python
import math

def count_k_permutations(lst, k):
    return math.factorial(len(lst)) // math.factorial(len(lst) - k)

# Test the function
print(count_k_permutations([1, 2, 3, 4], 2))
```

**Animal Joint Probability Problem**
A wildlife researcher is studying a forest habitat. They are particularly interested in the presence of two animal species:

Red foxes (F)
Barn owls (O)
Based on previous studies:

The probability of observing a red fox in a given area is 0.4
The probability of observing a barn owl in a given area is 0.3
The probability of observing both a red fox and a barn owl in the same area is 0.15
Calculate the following probabilities:

a) P(F  O): Probability of observing both a red fox and a barn owl in the same area

b) P(F  O): Probability of observing either a red fox or a barn owl (or both) in an area

c) P(F | O): Probability of observing a red fox, given that a barn owl has been observed

d) P(O | F): Probability of observing a barn owl, given that a red fox has been observed

e) P(F'  O'): Probability of observing neither a red fox nor a barn owl in an area

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib_venn import venn2

# Set default figure size for all plots in Jupyter
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100

# Given probabilities
P_F = 0.4  # Probability of observing a red fox
P_O = 0.3  # Probability of observing a barn owl
P_F_and_O = 0.15  # Probability of observing both a red fox and a barn owl

# Calculations
P_F_or_O = P_F + P_O - P_F_and_O
P_F_given_O = P_F_and_O / P_O
P_O_given_F = P_F_and_O / P_F
P_neither = 1 - P_F_or_O

# Print results
print(f"a) P(F  O) = {P_F_and_O:.4f}")
print(f"b) P(F  O) = {P_F_or_O:.4f}")
print(f"c) P(F | O) = {P_F_given_O:.4f}")
print(f"d) P(O | F) = {P_O_given_F:.4f}")
print(f"e) P(F'  O') = {P_neither:.4f}")

# Verification
total_probability = P_F_and_O + (P_F - P_F_and_O) + (P_O - P_F_and_O) + P_neither
print(f"\nSum of all probabilities: {total_probability:.4f}")

# Visualizations

# 1. Bar plot of probabilities
plt.figure(figsize=(14, 8))
probabilities = [P_F_and_O, P_F_or_O, P_F_given_O, P_O_given_F, P_neither]
labels = ['P(F  O)', 'P(F  O)', 'P(F | O)', 'P(O | F)', "P(F'  O')"]
plt.bar(labels, probabilities)
plt.title('Probabilities in the Animal Joint Probability Problem', fontsize=16)
plt.ylabel('Probability', fontsize=14)
plt.ylim(0, 1)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
for i, v in enumerate(probabilities):
    plt.text(i, v + 0.01, f'{v:.3f}', ha='center', fontsize=12)
plt.show()

# 2. Venn diagram
plt.figure(figsize=(12, 10))
v = venn2(subsets=(P_F - P_F_and_O, P_O - P_F_and_O, P_F_and_O),
          set_labels=('Red Fox', 'Barn Owl'))
plt.title('Venn Diagram of Red Fox and Barn Owl Observations', fontsize=16)
plt.text(0.5, -0.1, f"P(neither) = {P_neither:.3f}", ha='center', fontsize=14, transform=plt.gca().transAxes)
for text in v.set_labels:
    text.set_fontsize(14)
for text in v.subset_labels:
    if text is not None:
        text.set_fontsize(12)
plt.show()

# 3. Pie chart of all outcomes
plt.figure(figsize=(12, 12))
sizes = [P_F_and_O, P_F - P_F_and_O, P_O - P_F_and_O, P_neither]
labels = ['Both', 'Only Fox', 'Only Owl', 'Neither']
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, textprops={'fontsize': 14})
plt.axis('equal')
plt.title('Distribution of All Possible Outcomes', fontsize=16)
plt.show()
```
