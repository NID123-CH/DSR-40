```python
import numpy  as np
from sklearn . datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

def train_knn_model(X_train,y_train,n_neighbors):
    knn=KNeighborsClassifier(n_neighbors =n_neighbors)
    knn.fit(X_train,y_train)
    return knn

def evaluate_model(model,X_test,y_test):
    accuracy = model.score( X_test , y_test )
    print( "Model accuracy: ", accuracy)
    return accuracy


def plot_accuracies(neighbors,accuracies):
    plt.plot(neighbors ,accuracies,marker='o')
    plt.xlabel( 'Number of Neighbors' )
    plt.ylabel('Accuracy')
    plt.title( 'KNN Classifier Accuracy')
    plt.show( )

iris = load_iris()
X = iris.data
y = iris.target
x_train, x_test, y_train,y_test=train_test_split( X ,y ,test_size= 0.3 , random_state=42)
accuracies=[]
neighbors=range(1, 11)

for n in neighbors :
    model=train_knn_model( x_train , y_train ,n )
    accuracy=evaluate_model( model , x_test ,y_test )
    accuracies.append (accuracy)

plot_accuracies ( neighbors , accuracies )
```

```python
# prompt: Write a function where we can multiply 2 numbers together, and add any number to it.

def multiply_and_add(num1, num2, add_num):
  """
  Multiplies two numbers together and adds another number to the result.
  """
  product = num1 * num2
  result = product + add_num
  return result

# Example usage
num1 = 5
num2 = 3
add_num = 2
result = multiply_and_add(num1, num2, add_num)
print("Result:", result)
```

```python
# prompt: Write a function to sum all the numbers in a list.

def sum_list(numbers):
  """
  Calculates the sum of all numbers in a list.
  """
  total = 0
  for number in numbers:
    total += number
  return total
```

```python
# prompt: Write a function to multiply all the numbers in a list. Sample List : (8, 2, 3, -1, 7)

def multiply_list(numbers):
  """
  Multiplies all the numbers in a list.
  """
  result = 1
  for number in numbers:
    result *= number
  return result

# Sample list
numbers = (8, 2, 3, -1, 7)

# Multiply all numbers in the list
product = multiply_list(numbers)

# Print the result
print("Product:", product)
```

```python
# prompt: Write a function to check whether a number is in a given range.

def is_in_range(number, lower_bound, upper_bound):
  """
  Checks whether a number is in a given range.
  """
  if lower_bound <= number <= upper_bound:
    return True
  else:
    return False
```

```python
# prompt: Write a function  for  Sample List : [1,2,3,3,3,3,4,5] Unique List : [1, 2, 3, 4, 5]that takes a list and returns a new list with unique elements of the first list.

def unique_list(input_list):
  """
  Returns a new list with unique elements of the input list.
  """

  unique_elements = []
  for item in input_list:
    if item not in unique_elements:
      unique_elements.append(item)
  return unique_elements

  input_list = [1, 2, 3, 3, 3, 3, 4, 5]
  unique_list = unique_list(input_list)
  print("Unique List:", unique_list)
```

```python
import numpy as np

def generate_data_numpy(num_entries=100000):
    # Generate Pack Size between 6 and 24 (common pack sizes)
    pack_size = np.random.randint(6, 25, num_entries)

    # Generate Bottle Volume between 330 ml and 2000 ml
    bottle_volume = np.random.randint(330, 2001, num_entries)

    # Generate State Bottle Cost between $1 and $10
    state_bottle_cost = np.round(np.random.uniform(1, 10, num_entries), 2)

    # Generate Packs Sold between 1 and 500
    packs_sold = np.random.randint(1, 501, num_entries)

    # Calculate Sale = Packs Sold * Pack Size * State Bottle Cost
    sale = np.round(packs_sold * pack_size * state_bottle_cost, 2)

    # Stack all the variables into a 5x100000 array
    data = np.vstack((sale, pack_size, state_bottle_cost, packs_sold, bottle_volume))

    return data

# Generate data
amount =10_000
data = generate_data_numpy(100000)

# Check the shape of the data
print(data.shape)  # Should print (5, 100000)

# Example: print the first 5 columns of data
print(data[:, :5])
```

```python
def find_greatest_product(grid):
  """Finds the greatest product of four adjacent numbers in a grid.

  Args:
    grid: A 2D list representing the grid of numbers.

  Returns:
    The greatest product found.
  """

  max_product = 0

  # Iterate over each element in the grid
  for i in range(len(grid)):
    for j in range(len(grid[0])):
      # Check for horizontal products
      if j < len(grid[0]) - 3:
        product = grid[i][j] * grid[i][j+1] * grid[i][j+2] * grid[i][j+3]
        max_product = max(max_product, product)

      # Check for vertical products
      if i < len(grid) - 3:
        product = grid[i][j] * grid[i+1][j] * grid[i+2][j] * grid[i+3][j]
        max_product = max(max_product, product)


      # Check for diagonal products (down-right)
      if i < len(grid) - 3 and j < len(grid[0]) - 3:
        product = grid[i][j] * grid[i+1][j+1] * grid[i+2][j+2] * grid[i+3][j+3]
        max_product = max(max_product, product)

      # Check for diagonal products (up-right)
      if i >= 3 and j < len(grid[0]) - 3:
        product = grid[i][j] * grid[i-1][j+1] * grid[i-2][j+2] * grid[i-3][j+3]
        max_product = max(max_product, product)

  return max_product
```
